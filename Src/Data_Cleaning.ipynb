Handle missing data

```python
# check nan values
cols_with_nan = {}
for col in data.columns:
    nan_count = data[col].isnull().sum()
    if nan_count > 0:
        cols_with_nan[col] = nan_count

for col, count in cols_with_nan.items():
    print(f"{col} - {count}")
```

```python
# remove the columns that have the highest values of nan values and will not be used in this project
data = data.drop(['Renewable-electricity-generating-capacity-per-capita', 
                  'Financial flows to developing countries (US $)', 
                  'Renewables (% equivalent primary energy)'], axis=1)
```

```python
# remove columns that will not be used
data = data.drop(["Access to electricity (% of population)", "Access to clean fuels for cooking",
                  "Renewable energy share in the total final energy consumption (%)",
                  "Electricity from nuclear (TWh)", "Energy intensity level of primary energy (MJ/$2017 PPP GDP)", 
                  "Electricity from nuclear (TWh)"], axis=1)
```

```python
# remove necessary countries
countries_to_remove = ['Afghanistan', 'Aruba', 'Bahamas', 'Bermuda', 'Cayman Islands', 'Congo',
       'Czechia', 'Djibouti', 'Egypt', 'Eritrea', 'French Guiana', 'Gambia',
       'Kyrgyzstan', 'Liberia', 'Nauru', 'New Caledonia', 'Qatar', 'Saint Kitts and Nevis',
       'Saint Lucia', 'Saint Vincent and the Grenadines',
       'Sao Tome and Principe', 'Slovakia', 'Somalia', 'South Sudan',
       'Turkmenistan', 'Tuvalu', 'Yemen']
```

```python
data = data[~data['Entity'].isin(countries_to_remove)]
```

```python
# check again Value_co2_emissions_kt_by_country column
data[data["Value_co2_emissions_kt_by_country"].isnull()]["Entity"].nunique()
```

```python
# check nan values
cols_with_nan = {}
for col in data.columns:
    nan_count = data[col].isnull().sum()
    if nan_count > 0:
        cols_with_nan[col] = nan_count

for col, count in cols_with_nan.items():
    print(f"{col} - {count}")
```

```python
# Countries with the highest co2 emissions
average_co2_emission_by_country = data.groupby('Entity')['Value_co2_emissions_kt_by_country'].mean()
top_10 = average_co2_emission_by_country.nlargest(10)
```
